# 3.음식점 평가 데이터

```python
!pip install konlpy
from konlpy.tag import Okt
import re
```

```python
# 정규 표현식 함수 정의

import re

def apply_regular_expression(text):
    hangul = re.compile('[^ ㄱ-ㅣ 가-힣]')  # 한글 추출 규칙: 띄어 쓰기(1 개)를 포함한 한글
    result = hangul.sub('', text)  # 위에 설정한 "hangul"규칙을 "text"에 적용(.sub)시킴
    return result
```

```python
from collections import Counter
```

```python
okt = Okt()
okt.nouns(apply_regular_expression(review_score['text'][0]))
```

##3-1.긍정 부정 나누기

```python
# 말뭉치 생성
pos_corpus = "".join(pos_review['text'].tolist())
neg_corpus = "".join(neg_review['text'].tolist())
pos_corpus
```

```python
# 정규 표현식 적용
apply_regular_expression(pos_corpus)
```

```python
# 정규 표현식 적용
apply_regular_expression(neg_corpus)
```

```python
# 전체 말뭉치(corpus)에서 명사 형태소 추출
pos_nouns = okt.nouns(apply_regular_expression(pos_corpus))
neg_nouns = okt.nouns(apply_regular_expression(neg_corpus))
print(pos_nouns)
```

```python
# 빈도 탐색
pos_counter = Counter(pos_nouns)
neg_counter = Counter(neg_nouns)
pos_counter.most_common(10)
```

##3-2.긍정 부정 리뷰 그래프

```python
max = 20
pos_top_20 = {}
for word, counts in pos_counter.most_common(max):
    pos_top_20[word] = counts
    print(f'{word} : {counts}')
```

```python
max = 20
neg_top_20 = {}
for word, counts in neg_counter.most_common(max):
    neg_top_20[word] = counts
    print(f'{word} : {counts}')
```

```python
plt.figure(figsize=(10, 5))
plt.title('긍정 리뷰의 단어 상위 (%d개)'% max, fontsize=17)
plt.ylabel('단어의 빈도수')
plt.xticks(rotation=70)
for key, value in pos_top_20.items():
    plt.bar(key, value, color='lightgrey')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.title('부정 리뷰의 단어 상위 (%d개)'% max, fontsize=17)
plt.ylabel('단어의 빈도수')
plt.xticks(rotation=70)
for key, value in neg_top_20.items():
    plt.bar(key, value, color='lightgrey')
plt.show()
```

```python
%matplotlib inline

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
!pip install sklearn
import sklearn
from sklearn.feature_extraction.text import CountVectorizer
```

전체데이터(review_score) 처리

##4-1.불용어 사전

```python
stopwords = pd.read_csv("https://raw.githubusercontent.com/yoonkt200/FastCampusDataset/master/korean_stopwords.txt").values.tolist()
stopwords[:10]
```

```python
def text_cleaning(text):
    hangul = re.compile('[^ ㄱ-ㅣ 가-힣]')  # 정규 표현식 처리
    result = hangul.sub('', text)
    okt = Okt()  # 형태소 추출
    nouns = okt.nouns(result)
    nouns = [x for x in nouns if len(x) > 1]  # 한글자 키워드 제거
    nouns = [x for x in nouns if x not in stopwords]  # 불용어 제거
    return nouns

vectorizer = CountVectorizer()
vect = CountVectorizer(tokenizer = lambda x: text_cleaning(x))
bow_vect = vect.fit_transform(review_score['text'].tolist())
word_list = vect.get_feature_names_out()
count_list = bow_vect.toarray().sum(axis=0)
```

##4-2.TF-IDF

```python
from sklearn.feature_extraction.text import TfidfTransformer

tfidf_vectorizer = TfidfTransformer()
tf_idf_vect = tfidf_vectorizer.fit_transform(bow_vect)
```

```python
print(tf_idf_vect.shape)
```

```python
# 첫 번째 리뷰에서의 단어 중요도(TF-IDF 값) -- 0이 아닌 것만 출력
print(tf_idf_vect[0])
```

```python
# 첫 번째 리뷰에서 모든 단어의 중요도 -- 0인 값까지 포함
print(tf_idf_vect[0].toarray().shape)
print(tf_idf_vect[0].toarray())
```

#5.긍정 부정 예측

```python
from sklearn.model_selection import train_test_split

x = tf_idf_vect
y = review_score['y']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state=1)
```

```python
x_train.shape, y_train.shape
```

```python
x_test.shape, y_test.shape
```

##5-1.모델학습

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# fit in training set
lr = LogisticRegression(random_state = 0)
lr.fit(x_train, y_train)

# predict in test set
y_pred = lr.predict(x_test)
```

```python
# classification result for test set

print('accuracy: %.2f' % accuracy_score(y_test, y_pred))
print('precision: %.2f' % precision_score(y_test, y_pred))
print('recall: %.2f' % recall_score(y_test, y_pred))
print('F1: %.2f' % f1_score(y_test, y_pred))
```
